```python
def get_iso_code(country_name):
    try:
        url = 'https://restcountries.eu/rest/v2/name/' + country_name
        r = requests.get(url)
        data = r.json ()
        return data[0]['alpha3Code']
    except KeyError:
        print('[ERROR] cannot get code for', country_name)
        return None
    # print(country, data[0]['alpha3Code'])
    
get_iso_code('Thailand')
```

```python
def get_indicator(country_code, indicator):
    head = 'http://api.worldbank.org/v2/country/'
    middle= '/indicator/'
    tail= '?format=json'
    url= head+country_code+middle+indicator+tail
    r = requests.get(url)
    try:
        data= r.json()
        for i in data[1]:
                if i['value'] is None:
                    continue
                else:
                    return i['value']
    except KeyError:
        return None
    except IndexError:
        return None

get_indicator('CHN','GC.TAX.IMPT.ZS')
```
